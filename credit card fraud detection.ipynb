```python
np.random.seed(0)

all_independent_variables = [c for c in raw_df.columns if c not in ['Class','Time']]
train_X,test_X,train_y,test_y = train_test_split(raw_df[all_independent_variables],raw_df['Class'],test_size = .3)

eda_data = raw_df.iloc[train_X.index]
```

```python
eda_data.describe()
```

```python
corr = eda_data.corr()
corr.style.background_gradient()
```

```python
def get_metrics(pred,act,is_train = True):
    """
    pred: Predicted values
    act: actual values
    https://www.youtube.com/watch?v=yIYKR4sgzI8
    """
    acc = sum(pred == act)/act.shape[0]
    recall = sum(pred[act == 1])/act[act == 1].shape[0]
    precision = sum(pred[pred == 1] == act[pred == 1])/pred[pred == 1].shape[0]
    prefix = 'Train' if is_train else 'Test'
    print(f'{prefix} Acc: {acc:.2%}   Recall: {recall:.2%}   Precision: {precision:.2%}')
```

```python
corr_class = corr['Class'].loc[all_independent_variables].sort_values(key=lambda x:abs(x),ascending=False)
independent_vars = corr_class.index[[0]]

train_X_model = train_X[independent_vars]
test_X_model = test_X[independent_vars]

logreg = LogisticRegression()
logreg.fit(train_X_model,train_y)

train_y_pred = logreg.predict(train_X_model)
test_y_pred = logreg.predict(test_X_model)

get_metrics(train_y_pred,train_y,is_train = True)
get_metrics(test_y_pred,test_y,is_train = False)
```

```python
independent_vars = corr_class.index[:3]

train_X_model = train_X[independent_vars]
test_X_model = test_X[independent_vars]

logreg = LogisticRegression()
logreg.fit(train_X_model,train_y)

train_y_pred = logreg.predict(train_X_model)
test_y_pred = logreg.predict(test_X_model)

get_metrics(train_y_pred,train_y,is_train = True)
get_metrics(test_y_pred,test_y,is_train = False)
```
